```python
## ## Displaying the list of items availabe in store with quantity and cost
print("Welcome to Cart")
print()
item = ['Biscuits:a pack of 20', 'Cereals:pounds\pack', 'Chicken:1pound\pack','Oats:4pounds\pack','Rice:2pounds\pack']
quantity =[5,10,20,20,30]
cost = [2.50,3.50,13.30,3.90,4.90]
print("S.No\t","\titem\t\t",  "   quantity\t","   cost/item\t")
for i in range(len(item)):
    print(str(i+1) + '.\t' + item[i], quantity[i], cost[i],sep='\t\t')
def calculate_delivery_charge(distance):
    if distance <=5:
        return 10
    elif distance >5 and distance <=10:
        return 12
    elif distance >10 and distance <=15:
        return 14
    else:
        return 20
#asing customer to enter the details
def get_customer_details():
    customer_name = input("Enter name : ")
    delivery_address = input("Enter the address :")
    distance = float(input("Enter the distance in miles from the store to your address :"))
    return customer_name,delivery_address,distance
while True:
    order = []
    order.append(int(input("How many packets of biscuits are required? ")))
    order.append(int(input("How many packets of cereals are required? ")))
    order.append(int(input("How many packets of frozen chicken are required? ")))
    order.append(int(input("How many packets of oats are required? ")))
    order.append(int(input("How many packets of rice are required? ")))
    res = []
    for i,j in zip(quantity, order):
        res.append(i-j)
        
    for i in res:
        if i<=0:
            print("Item out of Stock",item[i])
        exit()
            
    customer_name,delivery_address,distance=get_customer_details()
    total_cost = sum([order[i]*cost[i] for i in range(len(order))]) 
    
    print("\t\tBill")
    print("S.No\t\t","Item\t\t\t  ","Quantity\t\t",     "Total Cost\t\t\t")
    for i in range(len(item)):
        if order[i] > 0:
            print(str(i+1) + '.\t' +   item[i],     order[i],order[i]*cost[i],sep='\t\t\t')
    print("Total Item Cost : $",total_cost)
    
    delivery_charge = calculate_delivery_charge(distance)
    print("Total Bill Amount : $",total_cost+delivery_charge)
    
    print("Name :",customer_name)
    print("Address :",delivery_address)  
    print()
    
    quantity=res
    
    if input("Do you want to continue shopping?(yes/no)").lower()!="yes":
        break
```
